# Curso de estadistica y probabilidad
## Probabilidad
Ciencia que estudia fenomenos aleatorios (aquel que tiene más de un posible resultado y este está fuera de nuestro control) y será el soporte de la estadistica. También existen los fenomenos deterministas que son los fenomenos en los que conocemos su resultado final)
## Estadistica
Ciencia que se encarga de recabar, organizar, presentar y analizar datos numericos para predecir situaciones futuras.
### Población
Aquel conjunto de aquellos elementos que comparten una caracteristica en comun.
### Investigación estadistica
Pasos que nos permiten saber las caracteristicas de alguna población en particular.
#### Problematicas al estudiar una población
1 - Una población puede ser muy grande. <br>
2 - Las poblaciones están localizadas el sitios inaccesibles. <br>
3 - Compuesta por elementos que no esten a nuestra disposición.
### Muestras
Sector de una población en el cual se pretende replicar el comportamiento de una población.

## Variable
Característica de una población que nos interesa estudiar. Siempre toma valores diferentes
### Variabilidad
Característica que tiene una población de tomar diferentes valores de elemento a elemento. (se debe al azar)
## Tipos
<ul>
    <li>Cuantitativas o númericas
        <ul>
            <li>Discretas: Se pueden contar</li>
            <li>Continuas: Se pueden medir</li>
        </ul>
    </li>
    <li>Cualitativas o categoricas
        <ul>
            <li>Ordinales: Criterio unico de orden. (por ejemplo los nombres de los meses, iniciamos en enero terminamos en diciembre)</li>
            <li>Nominales: No nos presentan un criterio unico para ordenarla. (por ejemplo los colores podemos ordenarlos de acuerdo al número de letras que podamos tener en la palabra o la preferencia de las personas encuestadas.)</li>
        </ul>
    </li>
<ul>

## Muestreo
Para estudiar una población tomamos una muestra de esta y sacamos conclusiones a partir de la información que esta nos genere.
<ul>
    <li>Debemos tener muy bien definida la variable de estudio.</li>
    <li>Escoger diferentes sectores de la población que puedan reresentar los datos de toda la población.</li>
    <li>Importante saber donde y cómo vamos a levantar los datos.</li>
</ul>

### Muestreo aleatorio simple
Es un tipo de muestreo se busca garantizar que cada elemento de la población tenga exactamente la misma probabilidad de ser seleccionado para la muestra.
### Tamaño de la muestra
Si la muestra es muy pequeña puede no ser representativa, en cambio muestras muuy grandes pueden llevar a costos muy elevados al hacer el estudio estadistico.

## Obtención de información estadistica
### Aplicación de un instrumento
<ul>
    <li>Cuestionario: tener en cuenta que las preguntas deben ser claras y no estar sujetas a interpretaciones, el tiempo, el espacio geografico, etc. Para tomar medidas debemos tomar en consideración el instrumento con el que se va a tomar mediciones </li>
</ul>

### Muestreo historico
Se puede recurrir a fuentes historicas de datos que se hayan levantado anteriormente.
### Población hipotetica
Se busca medir la influencia que puede tener una variable sobre una población hipotetica. (Ejemplo: cuando se trata de revisar la eficacia de un medicamento sobre una serie de pacientes, se genera información en lugar de levantarla.)

#### Censo
Aplicación de un instrumento para obtener información del total de la población.
#### Encuesta
Aplicación de un instrumento para obtener información sobre una muestra de la población.

## Gráficas

### Histograma
Nos va a permitir ver como es la frecuencia como son las frecuencias para cada variable y también como es la distribución de los datos

```python
import pandas as pd
a = pd.DataFrame()
a["altura"] = [50,50,50,50,50,50,55,55,60,60,80,80,100,100,100,108,118.5,318]
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
a.altura.plot.hist(edgecolor="black")
```

```python
a = pd.DataFrame()
a["altura"] = [50,55,60,80,100,108,118.5,318]
a["frecuencia"] = [6,2,2,2,3,1,1,1]
plt.scatter(a.altura,a.frecuencia,c="red")
plt.plot(a.altura,a.frecuencia)
```

### Ojiva

Esta gráfica nos muestra el crecimiento de la muestra ordenada

```python
lista = []
for i in range(len(a.frecuencia)):
    if i == 0:
        lista.append(a.frecuencia[i])
    else:
        lista.append(a.frecuencia[i]+lista[i-1])
a["frecuencia_acumulada"] = lista
```

```python
plt.scatter(a.altura, a.frecuencia_acumulada, c="red")
plt.plot(a.altura, a.frecuencia_acumulada)
```

### Gráficando frecuencias relativas
#### Ojiva porcentual

```python
import numpy as np
i = 0
a["frecuencia_relativa"] = [valor/np.sum(a.frecuencia) for valor in a.frecuencia]
lista = []
for i in range(len(a.frecuencia_relativa)):
    if i == 0:
        lista.append(a.frecuencia_relativa[i])
    else:
        lista.append(a.frecuencia_relativa[i] + lista[i-1])
a["frecuencia_relativa_acumulada"] = lista
a.head(10)
```

```python
plt.scatter(a.altura,a.frecuencia_relativa_acumulada,color="red")
plt.plot(a.altura,a.frecuencia_relativa_acumulada)
```
