## Matriz de dispersión

```python
from pandas.plotting import scatter_matrix
sm = scatter_matrix(df,alpha=0.2,diagonal="kde")
```

Podemos girar todas estos mensajes para una mejor visualización

```python
sm = scatter_matrix(df,alpha=0.2,figsize=(14,14),diagonal="kde")

#cambiando el nivel de rotacion
[s.xaxis.label.set_rotation(45) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

#Cambiando las coordenadas de cada elemento en los ejes
[s.get_yaxis().set_label_coords(-1,0.5) for s in sm.reshape(-1)]
[s.get_xaxis().set_label_coords(-0.2,0) for s in sm.reshape(-1)]

#Escondiendo los valores de los ejes
[s.set_xticks(()) for s in sm.reshape(-1)]
[s.set_yticks(()) for s in sm.reshape(-1)]
```

Podemos mirar las distribuciones de las variables haciendo uso de los "kde" y los histogramas

```python
valor_por_ciudad = df.groupby("CIUDAD").VALOR_MEDIANO.mean()
valor_por_ciudad.head()
```

## Diagrama de barras
Podemos representar variables categoricas con este diagrama.

```python
valor_por_ciudad.head(10).plot.barh()
```

## gráficos de linea
Se utilizan cuando se quieren mostrar tendencias temporales

```python
df.groupby("RATIO_PROFESORES").VALOR_MEDIANO.mean().plot.line()
```
