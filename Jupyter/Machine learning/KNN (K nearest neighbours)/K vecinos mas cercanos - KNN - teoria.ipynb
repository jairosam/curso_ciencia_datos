# K Nearest neighbours (K vecinos más cercanos) KNN

Es un problema de clasificacion que predice de acuerdo a los vecinos más proximos a una observación

Lo que hace el algoritmo es medir la distancia euclidiana a una observación aplicando el teorema de pitagoras con la formula que aparece en la imagen

Para problemas de clasificación o de regresión funciona de maneras distintas.

## Clasificación

Observa los vecinos más cercanos y clasifica de acuerdo a la cantidad de vecinos mayoritarios pertenecientes a una clase

## Regresión

Para problemas de regresión debemos hacer la media de los puntos más cercanos

## Pros:

* Fácil de implementar y de entender
* Fácil de modificar (2 hiperparametros, k y la metrica de la distancia)

## Contras: 

* Limitaciones al generalizar en datasets con alta dimensionalidad (muchas variables)
* Computacionalmente intensivo para predecir en datasets grandes (hay que calcular la distancia de la nueva observación a cada punto existente)
* No funciona bien con variables categoricas (la distancia entre ellas no es fácil de medir)
