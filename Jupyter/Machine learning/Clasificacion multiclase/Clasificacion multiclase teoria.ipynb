Los problemas que se haban manejado son de clasificacin binaria [0,1] o [verdadero, false], sin embargo la mayor parte de los problemas que se trabajan en la realidad son de clasificacin multiclase, por tanto para esto existen dos opciones

* Utilizar modelos que soportan multiples clases.
* Utilizar modelos que soportan clasificacin binaria con estrategias multiclase.
    * One-Versus-All
    * All-Versus-All

# One-Versus-All

Suponiendo que tenemos diferentes grupos **(manzanas, pias, bananas)**. Por tanto al generar los modelos se generarian 3 lineas rectas como se ven en las grficas. En el caso de la linea roja que describe las **manzanas** el modelo sabra que lo que est a la izquierda son **manzanas** y lo que esta a la derecha pues no lo son, de igual manera para las **bananas** y las **pias**

Finalmente se crea una distribucin de todo el espacio que separa las 3 categorias

# All-versus-All

En este tipo de modelos lo que hacemos es generar modelos que comparen dos categorias (manzanas vs bananas, manzanas vs naranja, naranjas vs bananas)

Al llegar una nueva observacin cada modelo determina que puede ser la entrada y luego se produce algn tipo de medida como una media ponderada para determinar que es la entrada

## Ventajas y desventajas
* One-Versus-all
    * Ventajas
        * puedes entrenar n modelos.
    * Desventajas
        * Los modelos van a estar desbalanceados, por ejemplo la clase de las manzanas representa 1/3 de la poblacin mientras que la clase no manzanas representa 2/3

* All-Versus-All
    * Ventajas
        * Como se compara una categoria con otra los datos de entrenamiento son ms balanceados. Por tanto los errores individuales del modelo van a ser ms pequeos.
    * Desventajas
        * Tiene muchos ms modelos y eso puede introducir ms error
