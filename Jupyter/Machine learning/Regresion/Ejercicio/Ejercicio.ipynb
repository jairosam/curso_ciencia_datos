```python
import pandas as pd
from sklearn.linear_model import LinearRegression as lr
```

```python
peliculas = pd.read_csv("movies.csv")
peliculas.head()
```

```python
peliculas = peliculas.select_dtypes(["int","float"])
peliculas = peliculas.fillna(0)
peliculas.head()
```

```python
modelo = lr()
```

```python
variable_dependiente = "ventas"
variables_independientes = peliculas.drop(columns=variable_dependiente).columns
```

```python
modelo.fit(X=peliculas[variables_independientes], y=peliculas[variable_dependiente])
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
peliculas["ventas_pred"] = modelo.predict(peliculas[variables_independientes])
```

```python
peliculas["ventas_pred"] = round(peliculas["ventas_pred"],3)
```
