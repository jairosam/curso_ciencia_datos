## Intro a aprendizaje supervisado



Representamos mediante un modelo de caja negra como funciona el aprendizaje supervisado, donde tenemos una nueva observación que se somete a un modelo predictivo y obtenemos una predicción.



Si tenemos un dataset que sometemos a un algoritmo de aprendizaje, esto producira un modelo predictivo



En la anterior imagen se muestra la estructura de un algoritmo de aprendizaje. 
1. Inicialización de parametros
2. Calculo de error
3. Mejora de parametros
4. Modelo entrenado

el calculo de error y la mejora de parametros se encuentra en un ciclo dado que es un proceso que se realiza varias veces para minimizar el error.



El dataset se debe dividir en dos partes, una para el entrenamiento y otra para las pruebas.

Esto se hace con el objetivo de evitar dos problemas:

## Overfitting (Sobreajuste)
Consiste en que se genera un modelo muy complejo, lo cual en otras palabras hace que el modelo se "memorice" la información del dataset por tanto las predicciones van a ser muy acertadas, sin embargo este modelo captura unicamente los datos del dataset y no obtiene ningún tipo de inteligencia sobre que variables influyen en el fenomeno.

## Underfitting (Sobregeneralización)
indica que el modelo genera patrones demasiado simples, por tanto no captura las caracteristicas del fenomeno
