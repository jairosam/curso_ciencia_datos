# Estadistica Bayesiana

## ¿Qué es una probabilidad?

la probabilidad es una medida de la certidumbre asociada a un suceso o evento futuro y se expresa con un numero entre 0 y 1

### Espacio muestral (S)

Lista de todos los posibles resultados de un experimento

Experimento de lanzar un dado
S = {1,2,3,4,5,6}

### Regla de Laplace 

Para obtener la probabilidad de un evento aplicamos esta regla

$$A=1$$

$$P(A)=\dfrac{Casos \ Favorables}{Casos \ posibles}$$

Esta si aplica solo si los eventos son equiprobables, es decir que sus probabilidades son iguales

## Eventos Complementarios

Los eventos complementarios son los eventos que completan el espacio muestral

$$P(A^c) = 1 - P(A)$$

## Probabilidad conjunta

Probabilidad de que dos eventos se den al mismo tiempo o simultaneamente **$P(A \cap B)$**

## Unión de probabilidades

La probabilidad de que suceda un evento u otro.

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

## Probabilidad condicional

Nos indica la probabilidad de que suceda un evento dado que ya sucedio otro

$$P(A|B) = \dfrac{P(A \cap B)}{P(B)}$$

## Eventos independientes

Son aquellos en los que la ocurrencia de uno no afecta la ocurrencia del otro

$$P(A \cap B) = P(A)P(B)$$

$$P(B|A) = \dfrac{P(B \cap A)}{P(A)} = \dfrac{P(A)P(B)}{P(A)} = P(B)$$

# Teorema de Bayes

El teorema establece que: 

$$P(A|B) = \dfrac{P(B|A)P(A)}{P(B)}$$

# Clasificadores Naive Bayesianos (Naive Bayes Classifier)

### Inferencia Bayesiana para clasificar

$$P(Clase_c|\{x_i\}) = \dfrac{P(\{x_i\}|Clase_c)P(Clase_c)}{P(\{x_i\})}$$

$$P(\{x_i\}|Clase_c) = P((x_1,x_2,...,x_n)|Clase_c)$$

Esto es posible calcular. sin embargo para poder computar de forma estadisticamente significativa la ocurrencia de dicha observación en la clase_c se necesitaria una cantidad de datos enormes para poder llegar a un resultado. 

Sin embargo se asumimos que las variables $x_i$ son independientes las unas de las otras

$$P(\{x_i \}|Clase_c) \approx P(x_1|Clase_c)*P(x_2|Clase_c)*...*P(x_n|Clase_c)$$

El clasificador Naive bayesiano funciona aprendiendo las probabilidades de ver cada una de las variables de las observaciones de los datos de entrenamiento para cada clase, y usando el teorema de bayes para predecir $P(Clase_c|observacion_i)$

#### Ejemplo de aplicaciones NBC - Noticias

Supongamos que queremos clasificar noticias en función de sus titulares como noticias de **Política** o **No Política** 

Tenemos los siguientes titulares 

Titular T1, Clase **Política**: "El Gobierno aprueba en madrid el presupuesto"

Titular T2, Clase **No Política**: "El Real Madrid vence al Barcelona"

Titular T3, Clase **Política**: "El gobierno redacta el Real Decreto 2017/99"

Titular T4, Clase **No Política**: "El Barcelona vence al Manchester"

Lo primero que debemos hacer es vectorizar

Para el titular T5 funciona de la siguiente manera, recordando que la suposición es que todos los eventos son independientes

$$P(C_p|T5) = P(gobierno|C_p) * P(madrid|C_p) * P(C_p) = 1 * 0.5 * 0.5 = 0.25$$
$$P(C_{np}|T5) = P(gobierno|C_{np}) * P(madrid|C_{np}) * P(C_{np}) = 0 * 0.5 * 0.5 = 0$$

Por tanto el titular T5 será clasificado como de la clase **Política**
